# Testing Custom Model

```python
from ultralytics import YOLO
```

```python
customModel = YOLO("CustomModel.pt")  # load a custom pretrained model
```

```python
customModel
```

```python
results2 = customModel.predict(source="datasets//images//thankyou_20240119_22_24_36_Pro.jpg")
```

```python
for detection in results2[0]: # Loop over each detection in the first image of the batch

    class_index = detection.boxes # Get the class index of the detection
    class_name = detection.names  # Get the name of the detection
    
    tensor_idx = class_index.cls
    conf_score = class_index.conf
    print(class_name[ tensor_idx.item()], conf_score.item())
```

```python
from PIL import Image
for r in results2:
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.show()  # show image
    im.save('custom_results.jpg')  # save image
```

```python
import cv2

classes_detected = []
class_indxs = []
webcam = cv2.VideoCapture(0)
#results3 = customModel.predict(source="0", conf=0.85 ,show=True)

# Loop through the video frames
while webcam.isOpened():
    # Read a frame from the video
    success, frame = webcam.read()

    if success:
        # Run YOLOv8 inference on the frame
        results3 = customModel(frame)
        for detection in results3[0]: # Loop over each detection in the first image of the batch
            class_index = detection.boxes # Get the class index of the detection
            class_name = detection.names  # Get the name of the detection
            
            conf_score = class_index.conf
            if conf_score.item() > 0.8:
                tensor_idx = class_index.cls
                class_indxs.append(tensor_idx)
                classes_detected.append(class_name[ tensor_idx.item()])

        # Visualize the results on the frame
        annotated_frame = results3[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Custom Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
webcam.release()
cv2.destroyAllWindows()
```

```python
print(classes_detected)
```

```python
from pathlib import Path
import onnx
from onnx_tf.backend import prepare
```

```python
tf_model_dir="sign-lang"
```

```python
onnx_model = onnx.load("CustomModel.onnx")
tf_rep = prepare(onnx_model)
tf_rep.export_graph(tf_model_dir)
```

```python
tfjs_model_dir = f"{tf_model_dir}-tfjs-fp32"
tfjs_model_dir
```

```python
tfjs_convert_command = f"""tensorflowjs_converter
                 --input_format=tf_saved_model
                 --output_format=tfjs_graph_model
                 --signature_name=serving_default
                 --saved_model_tags=serve
                 "{tf_model_dir}"
                 "{tfjs_model_dir}"
                 """
tfjs_convert_command = " ".join(tfjs_convert_command.split())
tfjs_convert_command
```
